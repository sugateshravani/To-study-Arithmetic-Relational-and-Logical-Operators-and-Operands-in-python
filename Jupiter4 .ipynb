```python
sum = 0

for i in range(1, 11):
    if i % 2 != 0:
        sum = sum + i

print("Sum of odd numbers:", sum)
```

```python
str1 = input("Enter a string: ")
result = ""

for i in range(len(str1)):
    if i % 2 == 0:
        result = result + str1[i]

print("Result:", result)
```
